## 전복 데이터셋 활용

```python
##  abalone   미국식 [│æbə│loʊni]
##  rings : 나이테개수
##  Gender : 성별
##  length :  길이
##  height : 두께
##  Diameter : 직경
##  Whole weight : 전체무게
##  Shucked weight : 순살무게
##  Viscera weight : 내장무게
##  Shell weight : 껍질무게
```

```python
import pandas as pd

df=pd.read_csv('data/abalone.csv',encoding='utf-8')
df.info()
```

```python
df.isnull().sum()
```

```python
# 경고 메시지 무시
import warnings
warnings.filterwarnings(action="ignore")
```

```python
## 나이테 수에 따른 전복 개수
df.groupby('Rings').count()
```

```python
## 나이테 수에 따른 순살의 평균 무게
df.groupby('Rings')['Shucked weight'].mean()
```

```python
## 나이테 수에 따른 순살 무게, 전체 무게와 껍질의 평균 무게
df.groupby('Rings')['Shucked weight','Whole weight','Shell weight'].mean()
```

```python
## 성별 분포 그래프 (sns 그래프)

import seaborn as sns

sns.countplot(x = 'Gender', data = df, palette="Set3")
```

```python
##  히스토그램으로 속성별 분포 확인
df.hist(figsize=(20,10), grid=False, layout=(2, 4), bins = 30)
```

```python
## 성별에 따른 나이테 분포 (swarmplot 과 violinplot 으로 )

import matplotlib.pyplot as plt

sns.swarmplot(x = 'Gender', y = 'Rings', data = df, hue = 'Gender')
sns.violinplot(x = 'Gender', y = 'Rings', data =df)

plt.show()
```

```python
## 각 속성의 상관관계 (pairplot)

import numpy as np

num_features = df.select_dtypes(include=[np.number]).columns
sns.pairplot(df[num_features])
```

```python
sns.heatmap(df[num_features].corr(), annot=True)
```

```python
# 껍질무게 이상치 제거
var = 'Shell weight'
plt.scatter(x = df[var], y = df['Rings'],)
plt.grid(True)
```

```python
df.drop(df[(df['Shell weight']> 0.6) & (df['Rings'] < 25)].index, inplace=True)
df.drop(df[(df['Shell weight']<1.0) & (df['Rings'] > 25)].index, inplace=True)
```

```python
var = 'Shell weight'
plt.scatter(x = df[var], y = df['Rings'],)
plt.grid(True)
```

```python
# 내장무게 이상치 제거
var = 'Viscera weight'
plt.scatter(x = df[var], y = df['Rings'],)
plt.grid(True)
```

```python
df.drop(df[(df['Viscera weight']> 0.5) & (df['Rings'] < 20)].index, inplace=True)
df.drop(df[(df['Viscera weight']<0.5) & (df['Rings'] >= 25)].index, inplace=True)

var = 'Viscera weight'
plt.scatter(x = df[var], y = df['Rings'])
plt.grid(True)
```

```python
# 순살 무게 이상치 제거

var = 'Shucked weight'
plt.scatter(x = df[var], y = df['Rings'])
plt.grid(True)
```

```python
df.drop(df[(df['Shucked weight']>= 1) & (df['Rings'] < 20)].index, inplace=True)
df.drop(df[(df['Shucked weight']<1) & (df['Rings'] > 20)].index, inplace=True)

var = 'Shucked weight'
plt.scatter(x = df[var], y = df['Rings'])
plt.grid(True)
```

```python
# 전체 무게 이상치 제거
var = 'Whole weight'
plt.scatter(x = df[var], y = df['Rings'])
plt.grid(True)
```

```python
df.drop(df[(df['Whole weight']> 2.0) & (df['Rings'] <= 20)].index, inplace=True)

var = 'Whole weight'
plt.scatter(x = df[var], y = df['Rings'])
plt.grid(True)
```

```python
# 직경 이상치 제거
var = 'Diameter'
plt.scatter(x = df[var], y =df['Rings'])
plt.grid(True)
```

```python
df.drop(df[(df['Diameter']<0.1) & (df['Rings']< 2.5)].index, inplace=True)

var = 'Diameter'
plt.scatter(x = df[var], y =df['Rings'])
plt.grid(True)
```

```python
# 높이(두께) 이상치 제거
var = 'Height'
plt.scatter(x = df[var], y =df['Rings'])
plt.grid(True)
```

```python
df.drop(df[(df['Height']>0.4) & (df['Rings'] < 15)].index, inplace=True)

var = 'Height'
plt.scatter(x = df[var], y =df['Rings'])
plt.grid(True)
```

```python
# 길이 이상치
var = 'Length'
plt.scatter(x = df[var], y =df['Rings'])
plt.grid(True)

## 이상치 없음
```
