### price 결측치 수정

```python
flg_is_null = sales_data['item_price'].isnull() # True or False 로 된 List (F,T,T...)

# 결측치 존재 상품명 추출(중복제거, unique), 상품A,S....List
for trg in list (sales_data.loc[flg_is_null, 'item_name'].unique()):
    # 같은 상품명의 가격을 ( Maxium 가격) 가져옴
    price = sales_data.loc[(~flg_is_null) & (sales_data['item_name'] == trg), 'item_price'].max()
    # 가져온 금액으로 수정,
    sales_data['item_price'].loc[(flg_is_null) & ( sales_data['item_name'] == trg)] = price

sales_data
```

###고객 이름 수정

```python
customer_data['고객이름'] = customer_data['고객이름'].str.replace(" ","") #공백 없애기
customer_data['고객이름'] = customer_data['고객이름'].str.replace(" ","") #공백 없애기
customer_data.head()
```

```python
customer_data.dtypes
```

```python
# 숫자형태있는 곳을 True로
flg_is_serial = customer_data['등록일'].astype('str').str.isdigit()
flg_is_serial
```

```python
flg_is_serial.sum()
```

```python
# 숫자를 낭짜로 변환 ( 일자별로 Day 로 계산, 1900/01/01시작점으로)

fromSerial = pd.to_timedelta(customer_data.loc[flg_is_serial, '등록일']. astype('float'), unit = 'D' ) + pd.to_datetime('1900/01/01')
fromSerial
```

```python
# 문자형식의 날자 값을 날자 현식으로 변롼

fromString = pd.to_datetime(customer_data.loc[~flg_is_serial,'등록일'])
fromString                           
```

```python
customer_data['등록일'] = pd.concat([fromSerial, fromString])
customer_data
```

```python
customer_data['등록년월'] = customer_data['등록일'].dt.strftime('%Y%m')
customer_data
```

```python
result = customer_data.groupby('등록년월').count()['고객이름']
result
```
