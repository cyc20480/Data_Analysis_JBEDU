```python
G = nx.Graph()   ## 무방향
G.add_nodes_from([1,2,3,4,5]) # 노드 추가
G.add_edges_from([(1,2), (1,3), (2,4), (3,5), (4,5)]) # 엣지 추가
```

```python
# 최단경로찾기

start_node = 1
end_node = 5
shortest_path = nx.shortest_path(G, source = start_node, target = end_node)
print(f'최단경로:{shortest_path}')
```

### 기본 Network Graph_003

```python
G = nx.DiGraph()   ## 방향
edges = [(1,2), (1,3), (2,1), (2,3), (3,4), (4,5), (5,3), (5,1)]
G.add_edges_from(edges)

fig, ax = plt.subplots()
pos = nx.spring_layout(G) # Node 간의 힘을 가상으로 설정하여 배치
nx.draw(G, pos, with_labels = True, font_weight = 'bold', node_size = 1000, 
        node_color = 'skyblue', font_color = 'black', font_size = 12,
       edge_color = 'gray', linewidths = 1, alpha = 0.7, ax = ax )
plt.title('방향성 Graph 예제')
plt.show()
```

```python
# 페이지 링크 알고리듬으로 중요성 측정, 중요성 높은 노드 찾기

pagerank_values = nx.pagerank(G)

most_important_node = max(pagerank_values, key = pagerank_values.get)

print(f'가장 중요한 역할을 하는 노드 : { most_important_node}')
```

### 기본 Network Graph_004

```python
#Social Network Data Random 생성

np.random.seed(42) 
num_nodes = 10
network_data = np.random.randint(2, size = (num_nodes, num_nodes)) # 10행, 10열 2차원배열
np.fill_diagonal(network_data,0) #행렬의 대각선 자리는 0으로
```

```python
df_social_network = pd.DataFrame(network_data, columns=[f'node{i}' for i in range(1, num_nodes + 1)])
df_social_network.to_csv('social_network.csv', index = False)
df_social_network = pd.read_csv('social_network.csv')
```

```python
df_social_network
```

```python
G = nx.Graph()   
G.add_nodes_from(df_social_network.columns)

for i in range(len(df_social_network.index)):
    for j in range(len(df_social_network.columns)):
        if df_social_network.iloc[i,j] == 1:
            G.add_edge(df_social_network.columns[i], df_social_network.columns[j])
```

```python
#중요성 지표
centrality = nx.degree_centrality(G) # 연결선 갯수로 중요성 판단
# pagerank_values = nx.pagerank(G) # 연결노드와 노드의 중요성에 영향을 받음
```

```python
most_important_node = max(centrality, key = centrality.get)

fig, ax = plt.subplots()
pos = nx.spring_layout(G) # Node 간의 힘을 가상으로 설정하여 배치
nx.draw(G, pos, with_labels = True, font_weight = 'bold', node_size = 1000, 
        node_color = 'skyblue', font_color = 'black', font_size = 10,
       edge_color = 'gray', linewidths = 1, alpha = 0.7, ax = ax )
nx.draw_networkx_nodes(G, pos, nodelist = [most_important_node], node_size = 3000,
                       node_color = 'red' , ax = ax )
plt.title('Social Network Graph 및 중요한 역할을 하는 노드')
plt.show()

print(f'가장 중요한 역할을 하는 노드 : {most_important_node}' )

start_node = 'node1'
end_node = 'node9'
shortest_path = nx.shortest_path(G, source = start_node, target = end_node)
print(f' node1 -> node9 최단경로:{shortest_path}')
```

### 기본 Network Graph_005

```python
nodes = [ i  for i in range(1,11)]
edges = [(1,2), (1,3), (1,4), (2,5), (2,6), (3,7), (3,8), (4,9), (4,10)]

G = nx.Graph()
G.add_nodes_from(nodes)
G.add_edges_from(edges)
```

```python
pos = nx.circular_layout(G) # Node를 원형으로 배치
fig, ax = plt.subplots()
nx.draw(G, pos, with_labels = True, font_weight = 'bold', node_size = 300, 
        node_color = 'skyblue', font_color = 'black', font_size = 8,
       edge_color = 'gray', linewidths = 1, alpha = 0.7, ax = ax )
plt.title('새로운 예제 Graph ( 원형 Layout )')
plt.show()
```

### 기본 Network Graph_006

```python
# 방향성이 있는 그래프 생성
G = nx.DiGraph()

# 노드 추가 (노드에는 'label'이라는 속성 추가)
nodes = [(1, {'label': 'A'}), (2, {'label': 'B'}), (3, {'label': 'C'}), (4, {'label': 'D'})]
G.add_nodes_from(nodes)

# 엣지 추가 (엣지에는 'weight'라는 속성 추가)
edges = [(1, 2, {'weight': 3}), (2, 3, {'weight': 1}), (3, 4, {'weight': 2}), (4, 1, {'weight': 4})]
G.add_edges_from(edges)
```

```python
pos = nx.circular_layout(G) 
labels = nx.get_node_attributes(G, 'label')
weights = nx.get_edge_attributes(G, 'weight')

fig, ax = plt.subplots()
nx.draw(G, pos, with_labels = True, font_weight = 'bold', node_size = 700, 
        node_color = 'skyblue', font_color = 'black', font_size = 12,
       edge_color = 'gray', linewidths = 1, alpha = 0.7, ax = ax )
nx.draw_networkx_edge_labels(G, pos,edge_labels = weights ) #엣지에 가중치 추가
plt.title('방향성 Graph 예제')
plt.show()
```
