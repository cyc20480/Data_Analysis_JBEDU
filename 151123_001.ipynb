```python
import matplotlib.pyplot as plt

plt.rc('font', family='malgun gothic')
plt.rcParams['axes.unicode_minus'] = False

plt.figure(figsize=(15,10))
plt.bar(hat['hatchery'], hat['chick'], color= ('red','orange','yellow','green','blue','navy','purple'))
plt.title('부화장 별 병아리 부화 현황')
plt.xlabel('부화장')
plt.ylabel('부화 마리 수')
plt.show()
```

```python
import seaborn as sns

col7=sns.color_palette('Pastel2',7)

plt.figure(figsize=(10,7))
plt.bar(hat['hatchery'], hat['chick'], color= col7, edgecolor = 'black')
plt.title('부화장 별 병아리 부화 현황')
plt.xlabel('부화장')
plt.ylabel('부화 마리 수')
plt.show()
```

```python
def addtext(x,y):
    for i in range(len(x)):
        plt.text(i, y[i]+0.5, y[i], ha = 'center')
        
        
col7=sns.color_palette('Pastel2',7)

plt.figure(figsize=(10,7))
plt.bar(hat['hatchery'], hat['chick'], color= col7, edgecolor = 'black')
addtext(hat['hatchery'], hat['chick'])
plt.title('부화장 별 병아리 부화 현황')
plt.xlabel('부화장')
plt.ylabel('부화 마리 수')
plt.show()
```

```python
def addtext(x,y):
    for i in range(len(x)):
        plt.text(i, y[i]/2 , y[i], ha = 'center')
        
        
col7=sns.color_palette('Pastel2',7)

plt.figure(figsize=(10,7))
plt.bar(hat['hatchery'], hat['chick'], color= col7, edgecolor = 'black')
addtext(hat['hatchery'], hat['chick'])
plt.title('부화장 별 병아리 부화 현황')
plt.xlabel('부화장')
plt.ylabel('부화 마리 수')
plt.show()
```

```python
col7=sns.color_palette('Pastel2',7)

plt.figure(figsize=(10,7))
plt.bar(hat['hatchery'], hat['chick'], color= col7, edgecolor = 'black')
addtext(hat['hatchery'], hat['chick'])

plt.hlines(hat.chick.mean(), -1, 7, color ='red', linestyle = 'dashed')

plt.title('부화장 별 병아리 부화 현황')
plt.xlabel('부화장')
plt.ylabel('부화 마리 수')
plt.show()
```

```python
col7=sns.color_palette('Pastel2',7)

plt.figure(figsize=(10,10))
plt.pie(hat['chick'], labels=hat['hatchery'], autopct = '%.1f%%',colors=col7)
plt.show()


# https://seaborn.pydata.org/tutorial/color_palettes.html#
```
