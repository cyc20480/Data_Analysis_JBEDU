```python
summ = 0

for i in data:
    if ( i[-1] == '전표매입' ) & ( i[0].split("-")[1] == '10' ):
       summ += int(i[6])
print(summ)
```

```python
oct_sum = 0
nov_sum = 0
dec_sum = 0

for i in data:
    if ( i[-1] == '전표매입' ) & ( i[0].split("-")[1] == '10' ):
       oct_sum += int(i[6])
    elif ( i[-1] == '전표매입' ) & ( i[0].split("-")[1] == '11' ):
       nov_sum += int(i[6])
    elif ( i[-1] == '전표매입' ) & ( i[0].split("-")[1] == '12' ):
       dec_sum += int(i[6])
    
print(f' 10월 {oct_sum}, 11월 {nov_sum}, 12월 {dec_sum} 원 결제')
```

```python
oct_sum = 0
nov_sum = 0
dec_sum = 0

for i in data:
    if ( i[-1] == '전표매입' ):
              if ( i[0].split("-")[1] == '10' ):
                   oct_sum += int(i[6])
              elif ( i[0].split("-")[1] == '11' ):
                   nov_sum += int(i[6])
              elif  ( i[0].split("-")[1] == '12' ):
                   dec_sum += int(i[6])
    
print(f' 10월 {oct_sum}, 11월 {nov_sum}, 12월 {dec_sum} 원 결제')
```

```python
sum_mon = [0,0,0]

for i in data:
    if ( i[-1] == '전표매입' ):
        sum_mon[int(i[0].split("-")[1]) - 10] += int(i[6])
        
print(sum_mon)
print(f' 10월 {sum_mon[0]}, 11월 {sum_mon[1]}, 12월 {sum_mon[2]} 원 결제')
```

```python
sum_mon = [0 for j in range(3)]

for i in data:
    if ( i[-1] == '전표매입' ):
        sum_mon[int(i[0].split("-")[1]) - 10] += int(i[6])
        
print(sum_mon)
print(f' 10월 {sum_mon[0]}, 11월 {sum_mon[1]}, 12월 {sum_mon[2]} 원 결제')
```

```python
import matplotlib.pyplot as plt

sum_mon = [0 for j in range(3)]

for i in data:
    if ( i[-1] == '전표매입' ):
        sum_mon[int(i[0].split("-")[1]) - 10] += int(i[6])
        
print(sum_mon)
print(f' 10월 {sum_mon[0]}, 11월 {sum_mon[1]}, 12월 {sum_mon[2]} 원 결제')

plt.rc ('font', family="Malgun gothic")
plt.title(" 10 ~ 12 월 지출 현황 ")
plt.bar([' 10 월 ', ' 11 월 ', ' 12 월 '], sum_mon, color='royalblue')
plt.plot([' 10월 ',' 11월 ',' 12월 '], sum_mon, color='royalblue')
plt.show()
```

```python
import matplotlib.pyplot as plt

sum_mon_t = [0 for j in range(3)]

for i in data:
    if ( i[-1] == '전표매입' ) & ( '택시' in i[5] ):
        sum_mon_t[int(i[0].split("-")[1]) - 10] += int(i[6])
        
print(sum_mon)
print(f' 10월 {sum_mon_t[0]}, 11월 {sum_mon_t[1]}, 12월 {sum_mon_t[2]} 원 택시비 결제')

plt.rc ('font', family="Malgun gothic")
plt.title(" 10 ~ 12 월 택시비 지출 현황 ")
plt.bar([' 10월 ', ' 11월 ', ' 12월 '], sum_mon, color='royalblue', label='택시비 지출액')
plt.plot([' 10월 ',' 11월 ',' 12월 '], sum_mon, color='red', label='택시비 지출액')
plt.legend(loc="best", fontsize = 10)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc ('font', family="Malgun gothic")

singer=['A','B','C','D','E']
week1=[42,58,19,92,84]
week2=[53,52,48,98,73]

for i in range(len(week1)):
    week1[i] = -week1[i]

plt.title('오디션 프로그램 득표현황')
plt.barh(singer,week1, label='첫째주', color='hotpink')
plt.barh(singer,week2, label='둘째주', color='royalblue')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc ('font', family="Malgun gothic")
plt.rcParams['axes.unicode_minus']=False  # '- 표시

singer=['A','B','C','D','E']
week1=[42,58,19,92,84]
week2=[53,52,48,98,73]

for i in range(len(week1)):
    week1[i] = -week1[i]

plt.title('오디션 프로그램 득표현황')
plt.barh(singer,week1, label='첫째주', color='hotpink')
plt.barh(singer,week2, label='둘째주', color='royalblue')
plt.legend()
plt.show()
```
