###### 데이터준비

```python
import pandas as pd
```

```python
red_df = pd.read_csv('data/winequality-red.csv', sep = ';', header = 0, 
                                                                 engine = 'python')
red_df
```

```python
white_df = pd.read_csv('data/winequality-white.csv', sep = ';', header = 0, 
                                                                 engine = 'python')
white_df
```

```python
## fixed acidity (고정 산도): 와인의 산성 정도를 나타내는 특성
## volatile acidity (휘발성 산도): 와인의 휘발성 산성 성분의 양
## citric acid (구연산): 와인에 함유된 구연산의 양
## residual sugar (잔류 설탕): 설탕의 양
## chlorides (염화물): 와인의 소금 함유량
## free sulfur dioxide (자유 이산화황): 이산화황의 양
## total sulfur dioxide (총 이산화황)
## density (밀도): 와인의 밀도
## pH: 와인의 수소 이온 농도
## sulphates(황산염)
## alcohol (알코올): 와인의 알콜 함유량
## quality (품질): 와인의 품질을 나타내는 등급
```

```python
red_df.to_csv('data/winequality-red2.csv',index = False)
white_df.to_csv('data/winequality-white2.csv', index = False)
```

```python
from scipy import stats
from statsmodels.formula.api import ols, glm
```

```python
## 레드와인 샘플의 quality값만 찾아서 저장

red_wine_quality = wine.loc[wine['type']=='red' , 'quality'  ]
```

```python
red_wine_quality
```

```python
## 화이트와인 샘플의 quality값만 찾아서 저장

white_wine_quality = wine.loc[wine['type'] == 'white', 'quality']

white_wine_quality
```

```python
stats.ttest_ind(red_wine_quality, white_wine_quality)
```

###### ** 회귀분석 (ols 사용)

```python
Rformula = 'quality ~ fixed_acidity + volatile_acidity + citric_acid + \
            residual_sugar + chlorides + free_sulfur_dioxide + \
            total_sulfur_dioxide + density + pH + sulphates + alcohol'
```

```python
r_result = ols(Rformula, data = wine).fit()
```

```python
r_result.summary()
```

```python
## wine에서 quality와 type을 제외하고, 회귀분석모델에 사용할 독립변수만 추출하여 저장

sample1 = wine[wine.columns.difference(['quality', 'type'])]  

sample1
```

```python
sample1 = sample1[0:5][:]    ## 예측에 사용할 샘플 추출
sample1
```

```python
## regression_result의 예측함수 predict()에 적용하여 수행한 뒤 결과 예측값을 저장

sample1_predict = r_result.predict(sample1)  

sample1_predict
```

###### 부분 회귀 플롯

```python
## 독립변수가 2개 이상인 경우 부분 회귀 플롯으로 
### 하나의 독립변수가 종속 변수에 미치는 영향력을 시각화하여 결과를 분석할 수 있다 
```

```python
import statsmodels.api as sm 
```

```python
others = list(set(wine.columns).difference(set(["quality", "fixed_acidity"])))
## 부분회귀에 사용한 독립변수와 종속변수를 제외한 나머지 변수 이름을 리스트로 추출

others
```

```python
p, resids = sm.graphics.plot_partregress("quality", "fixed_acidity", others, 
                                                 data = wine, ret_coords = True)

## 나머지 변수는 고정, "fixed_acidity"가 종속변수 "quality", 에 미치는 영향에 대한 
## 부분회귀수행
```

```python
fig = plt.figure(figsize = (8, 13))
sm.graphics.plot_partregress_grid(r_result, fig = fig)
plt.show()
```
