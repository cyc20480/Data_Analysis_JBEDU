```python
import random
height,weight,score=[],[],[]
for i in range(100):
    height.append(random.randint(100,200))
    weight.append(random.randint(20,200))
    score.append(random.randint(1,100))
print("키       : ", height)
print("몸무게   : ", weight)
print("체육점수 : ", score)
plt.rc('font',family='malgun gothic')    
plt.title('키와 몸무게, 체육 점수의 상관관계')
plt.scatter(height,weight,s=score, c=score, cmap='Reds', alpha=0.5)  #색상기준을 height로(키가 작을수록 연하게 클수록 진하게), 컬러맵
plt.colorbar(label='체육 점수')
plt.show()
```

```python
## 여의도 한강공원
## 여의도 공원

lat1, long1 = 37.52862, 126.9331 # 여의도 한강공원
lat2, long2 = 37.52565, 126.9215 # 여의도공원

map_y = folium.Map([lat1, long1], zoom_start=15)
folium.Marker([lat1, long1], tooltip='여의도 한강공원', icon=folium.Icon(color='red', icon='bomb',prefix='fa' )).add_to(map_y)
folium.Marker([lat2, long2], tooltip='여의도공원', icon=folium.Icon(color='blue', icon='home')).add_to(map_y)
map_y


### 다양한 마커 아이콘
### https://fontawesome.com/v4.7.0/icons
### fontawesome의 아이콘을 사용할 때에는 prefix='fa' 옵션을 추가
```

```python
lat = [37.52862, 37.52565, 37.51888]                         # 위도 리스트  
long = [ 126.9331, 126.9215, 126.9219]                     # 경도 리스트
names = ['여의도 한강공원', '여의도공원', '샛강생태공원']         # 위치 이름 리스트
icons = ['automobile', 'balance-scale', 'ban']               # 아이콘 모양 리스트   
colors = ['red', 'blue', 'purple']                           # 아이콘 색상 리스트

map_y = folium.Map([lat[0], long[0]], zoom_start = 15)

for i in range(len(lat)) : 
    folium.Marker([lat[i], long[i]], tooltip=names[i],
                  icon=folium.Icon(color=colors[i], icon=icons[i], prefix='fa')).add_to(map_y)

map_y.save('map_01.html')
```

```python
## 대학교 정보 엑셀파일을 데이터프레임으로 변환
import pandas as pd
import folium

df=pd.read_excel('data/서울지역 대학교 위치.xlsx', engine='openpyxl')
df
```
