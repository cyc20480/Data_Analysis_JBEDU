```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
from konlpy.tag import Okt
from PIL import Image
import numpy as np
from konlpy.tag import Hannanum
```

```python
txt = open('Data/wordcloud_001.txt','rt',encoding = 'UTF-8').read()
```

```python
txt
```

```python
# txt 문서에서 명사만 추출

n1 = Hannanum().nouns(txt)
n1
```

```python
n1[14] = '재구매'
```

```python
n1[0:15]
```

```python
# 글자수가 2글자 이상만 사용

n2 = [ i for i in n1 if len(i) >= 2]
```

```python
n2
```

```python
# 명사별 빈도 추출

cnt = Counter(n2)
cnt
```

```python
cnt.most_common()[0:10] # 빈도수 많은 순으로 10개 표시
```

```python
# 워드클라우드용 폰트 설정

wcs = WordCloud(font_path = 'C:/Windows/Fonts/MalangmalangB.ttf', background_color = 'white', colormap='RdBu')
```

```python
##  https://matplotlib.org/stable/users/explain/colors/colormaps.html
```

```python
cloud = wcs.generate_from_frequencies(cnt)
```

```python
plt.figure(figsize = (10,7))
plt.axis('off')
plt.imshow(cloud)
plt.savefig('data/wordcloud_001_image.png')
```
