### 데이터 분석 기초

```python
import csv

f=open("./data/card.csv", encoding="utf-8" )
data=csv.reader(f)

next(data)   # 표의 헤더 제거 (제목 행)

data=list(data)
print(data)
```

```python
s=0
for i in data:
    if i[-1]=='전표매입':
        s=s+int(i[6])
    
print(s)
```

```python
s=0
for i in data:
    if (i[-1]=='전표매입') | (i[-1] == '부분취소'):
        s=s+int(i[6])
    
print(s)
```

```python
s=0
for i in data:
    if (i[-1]=='전표매입') | ( '부분' in i[-1]):
        s=s+int(i[6])
    
print(s)
```

```python
for i in data:
    print(f'{i[5]} 에서 {int(i[6])} 원 결제')
```

```python
for i in data:
    print(f'{i[0]} {i[5]} 에서 {int(i[6])} 원 결제')
```

```python
for i in data:
    print(f'{i[0].split(" ")[0]} {i[5]} 에서 {int(i[6])} 원 결제')
```

```python
for i in data:
    print(f'{i[0].split("-")[1]}월 {i[5]} 에서 {int(i[6])} 원 결제')
```

```python
summ = 0

for i in data:
    if ( i[-1] == '전표매입' ) & ( i[0].split("-")[1] == '10' ):
       summ += int(i[6])
print(summ)
```

```python
oct_sum = 0
nov_sum = 0
dec_sum = 0

for i in data:
    if ( i[-1] == '전표매입' ) & ( i[0].split("-")[1] == '10' ):
       oct_sum += int(i[6])
    elif ( i[-1] == '전표매입' ) & ( i[0].split("-")[1] == '11' ):
       nov_sum += int(i[6])
    elif ( i[-1] == '전표매입' ) & ( i[0].split("-")[1] == '12' ):
       dec_sum += int(i[6])
    
print(f' 10월 {oct_sum}, 11월 {nov_sum}, 12월 {dec_sum} 원 결제')
```

```python
oct_sum = 0
nov_sum = 0
dec_sum = 0

for i in data:
    if ( i[-1] == '전표매입' ):
              if ( i[0].split("-")[1] == '10' ):
                   oct_sum += int(i[6])
              elif ( i[0].split("-")[1] == '11' ):
                   nov_sum += int(i[6])
              elif  ( i[0].split("-")[1] == '12' ):
                   dec_sum += int(i[6])
    
print(f' 10월 {oct_sum}, 11월 {nov_sum}, 12월 {dec_sum} 원 결제')
```

```python
sum_mon = [0,0,0]

for i in data:
    if ( i[-1] == '전표매입' ):
        sum_mon[int(i[0].split("-")[1]) - 10] += int(i[6])
        
print(sum_mon)
print(f' 10월 {sum_mon[0]}, 11월 {sum_mon[1]}, 12월 {sum_mon[2]} 원 결제')
```

```python
sum_mon = [0 for j in range(3)]

for i in data:
    if ( i[-1] == '전표매입' ):
        sum_mon[int(i[0].split("-")[1]) - 10] += int(i[6])
        
print(sum_mon)
print(f' 10월 {sum_mon[0]}, 11월 {sum_mon[1]}, 12월 {sum_mon[2]} 원 결제')
```

```python
import matplotlib.pyplot as plt

sum_mon = [0 for j in range(3)]

for i in data:
    if ( i[-1] == '전표매입' ):
        sum_mon[int(i[0].split("-")[1]) - 10] += int(i[6])
        
print(sum_mon)
print(f' 10월 {sum_mon[0]}, 11월 {sum_mon[1]}, 12월 {sum_mon[2]} 원 결제')

plt.rc ('font', family="Malgun gothic")
plt.title(" 10 ~ 12 월 지출 현황 ")
plt.bar([' 10 월 ', ' 11 월 ', ' 12 월 '], sum_mon, color='royalblue')
plt.plot([' 10월 ',' 11월 ',' 12월 '], sum_mon, color='royalblue')
plt.show()
```
