```python
import matplotlib.pyplot as plt

plt.rc('font',family='malgun gothic')
plt.rcParams['axes.unicode_minus']=False  # '- 표시

import random
height,weight=[],[]

for i in range(100):
    height.append(random.randint(100,200))
    weight.append(random.randint(20,200))
     
plt.scatter(height,weight,c=height,cmap='RdPu')  #색상기준을 height로(키가 작을수록 연하게 클수록 진하게), 컬러맵
plt.show()

#컬러맵 : https://matplotlib.org/stable/tutorials/colors/colormaps.html
```

```python
import random
height,weight,score=[],[],[]
for i in range(100):
    height.append(random.randint(100,200))
    weight.append(random.randint(20,200))
    score.append(random.randint(1,100))
print("키       : ", height)
print("몸무게   : ", weight)
print("체육점수 : ", score)
plt.rc('font',family='malgun gothic')    
plt.title('키와 몸무게, 체육 점수의 상관관계')
plt.scatter(height,weight,s=score, c=score,cmap='Reds', alpha=0.5) 
#색상기준을 height로(키가 작을수록 연하게 클수록 진하게), 컬러맵 s= 원의 크기
plt.colorbar(label='체육 점수')
plt.show()
```

```python
import folium
lat1, long1 = 37.52862, 126.9331
lat2, long2 = 37.52562, 126.9215

map_y = folium.Map([lat1, long1], zoom_start=15) #기준 위도와 경도
folium.Marker([lat1,long1], tooltip = ' 여의도 한강 공원 ', icon=folium.Icon(color='red', icon='bomb', prefix='fa')).add_to(map_y)
folium.Marker([lat2,long2], tooltip = ' 여의도 공원 ',icon=folium.Icon(color='blue', icon='home')).add_to(map_y)
map_y
```

```python
lat = [37.52862, 37.52565, 37.51888]                         # 위도 리스트  
long = [ 126.9331, 126.9215, 126.9219]                     # 경도 리스트
names = ['여의도 한강공원', '여의도공원', '샛강생태공원']         # 위치 이름 리스트
icons = ['automobile', 'balance-scale', 'ban']               # 아이콘 모양 리스트   
colors = ['red', 'blue', 'purple']                           # 아이콘 색상 리스

map_y = folium.Map([lat[0], long[0]], zoom_start = 15)
for i in range(len(lat)) : 
    folium.Marker([lat[i], long[i]], tooltip=names[i],
                  icon=folium.Icon(color=colors[i], icon=icons[i], prefix='fa')).add_to(map_y)

map_y
```

```python
lat = [37.52862, 37.52565, 37.51888]                         # 위도 리스트  
long = [ 126.9331, 126.9215, 126.9219]                     # 경도 리스트
names = ['여의도 한강공원', '여의도공원', '샛강생태공원']         # 위치 이름 리스트
icons = ['automobile', 'balance-scale', 'ban']               # 아이콘 모양 리스트   
colors = ['red', 'blue', 'purple']                           # 아이콘 색상 리스

map_y = folium.Map([lat[0], long[0]], zoom_start = 15)
for i in range(len(lat)) : 
    folium.Marker([lat[i], long[i]], tooltip=names[i],
                  icon=folium.Icon(color=colors[i], icon=icons[i], prefix='fa')).add_to(map_y)

map_y.save('map_01.html')
map_y.save('map_01.jpg')
map_y
```

```python
##대학교 정보 엑셀 파일 을 데이터 프레임으로 변환
```

```python
import pandas as pd

df = pd.read_excel('data/서울지역 대학교 위치.xlsx', engine = 'openpyxl')
df
```

```python
seoul_map = folium.Map(location=[37.5662952,126.9779451], zoom_start=12)

for name, lat, long in zip(df.학교, df.위도, df.경도):
    folium.Marker([lat, long], tooltip=name).add_to(seoul_map)
    
seoul_map.save("Seoul_univ.html")
seoul_map
```
